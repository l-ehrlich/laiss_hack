# XMM-Newton Data Retrieval with Astroquery and TAP

This notebook demonstrates how to retrieve **XMM-Newton** (ESA's X-ray observatory) data using **astroquery** and the **Table Access Protocol (TAP)**. The XMM-Newton Science Archive (XSA) provides a TAP service that we can query with **ADQL** (Astronomical Data Query Language).

## 1. Setup and imports

```python
# pip install astroquery  # if not already installed
from astroquery.esa.xmm_newton import XMMNewton
from astropy.table import Table

def get_column_names(table_name):
    """Return the list of column names for a TAP table (e.g. 'xsa.v_public_observations')."""
    return list(XMMNewton.get_columns(table_name))
```

## 2. Explore TAP tables

List all available tables in the XSA TAP service. These are the tables you can query with ADQL.

```python
tables = XMMNewton.get_tables()
print(f"Found {len(tables)} tables:\n")
for t in tables:
    print(f"  {t}")
```

## 3. Inspect columns of a table

Get the columns of `xsa.v_public_observations` (public observations) so we know what we can select in ADQL queries.

```python
table_name = 'tap_schema.columns'  # or any table from section 2
columns = get_column_names(table_name)
print(f"Columns in {table_name}:")
for col in columns:
    print(f"  {col}")
```

## 4. Run an ADQL TAP query

Query the first 10 public observations. Results are returned as an Astropy `Table`.

```python
result = XMMNewton.query_xsa_tap(
    "SELECT TOP 10 * FROM xsa.v_public_observations",
    output_format='votable'
)
result
```

## 5. Query with filters (e.g., by target or coordinates)

Example: get observations for a specific target name and minimum duration.

```python
query = """
SELECT observation_id, target, ra, dec, duration, start_utc
FROM xsa.v_public_observations
WHERE target LIKE '%Crab%'
  AND duration > 10000
ORDER BY start_utc DESC
"""
crab_obs = XMMNewton.query_xsa_tap(query)
crab_obs
```

## 6. Save TAP results to file

You can write results to CSV or VOTable for later use.

```python
result = XMMNewton.query_xsa_tap(
    "SELECT TOP 5 observation_id, target, ra, dec, duration FROM xsa.v_public_observations",
    output_format='csv',
    output_file='xmm_obs_sample.csv'
)
print("Saved to xmm_obs_sample.csv")
result
```

## 7. EPIC metadata via TAP (by target name)

`get_epic_metadata` uses the XSA TAP tables under the hood and returns EPIC source metadata for a given target (by name or coordinates).

```python
epic_source, cat_4xmm, stack_4xmm, slew_source = XMMNewton.get_epic_metadata(
    target_name="4XMM J122934.7+015657"
)
print("EPIC source table:")
epic_source
```
